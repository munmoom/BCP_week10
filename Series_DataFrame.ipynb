```python
import pandas as pd
```

```python
lst = [4.1, 5.5, 9.0, 15.0, 19.0, 21.3]
sr = pd.Series(lst)
sr
```

```python
idx = ['1월','2월','3월','4월','5월','6월']
lst = [4.1, 5.5, 9.0, 15.0, 19.0, 21.3]
sr = pd.Series(lst, index = idx)
sr
```

```python
print(sr['4월'])
print(sr[['4월','6월']])  #리스트
print(sr['4월':'6월'])  #끝값 포함
print(sr[:'4월'])
print(sr['4월':])
print(sr[sr.index>'4월'])
print(sr[sr>15.0])
```

```python
area = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기', '강원', '충북', '충남','전북']
#13
pm25_avg=[20,13,15,20,19,16,14,18,20,14,19,23,23]
pm25_max=[22,18,18,30,23,20,16,19,27,18,26,29,30]
pm25_min=[16,10,11,12,15,12,11,17,13,9,14,19,14]

df=pd.DataFrame()
df['지역']=area
df['평균']=pm25_avg
df['최대']=pm25_max
df['최소']=pm25_min

df
```

```python
print(df.loc[1,'지역'])
print(df.loc[1:3, '지역':'최대'])
print(df.loc[[1,3], ['지역','최소']])
```

```python
print(df.iloc[1,0])
print(df.iloc[1:4, 0:3])  #범위지정 끝값포함 안됨
print(df.iloc[[1,3], [0,3]])
```

```python
print(df[df['지역'] == '부산'])
print(df[df['최대'] >= 25])
print(df[(df['최대'] >= 18) & (df['최소'] <= 10)])
```
